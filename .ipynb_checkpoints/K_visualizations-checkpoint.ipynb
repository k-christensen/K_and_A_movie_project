```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('merged.csv')
df
```

```python
ax = sns.boxplot(x="month", y="domestic_gross", data=df, palette="Set3").set(ylabel= "Domestic Gross", title = "Movie Success by Month")

plt.show()
```
