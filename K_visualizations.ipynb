```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('merged.csv')
df
```

```python
tick_label_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
ax = sns.boxplot(x="month", y="domestic_gross", data=df, palette="Set3").set(ylabel= "Domestic Gross", title = "Movie Success by Month", xticklabels = tick_label_list)

plt.show()
```

```python
ax = sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Action'] == 1], palette="Set3").set(ylabel= "Domestic Gross", title = "Action Movie Success by Month")
plt.savefig('movie_boxplot.png')
plt.show()
```
