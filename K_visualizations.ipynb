```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('merged.csv')
df.head()
```

```python
tick_label_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.figure(figsize=(12, 6))
sns.set_context("talk")
ax = sns.boxplot(x="month", y="domestic_gross", data=df, palette="Set3").set(ylabel= "Domestic Gross", title = "Movie Success by Month", xticklabels = tick_label_list)
plt.savefig('movie_boxplot.png')
plt.show()
```

```python
nested_list_genres = list(df['genres'])

genre_list = []
for sublist in nested_list_genres:
    for item in sublist:
        genre_list.append(item)

def unique(lst):
    unique_list = []
    if item not in unique_list:
        unique_list.append(item)
    return unique_list

unique_genres = unique(nested_list_genres)
```

```python
genre_data = df.iloc[:, -21:-1]
genre_data['domestic_gross'] = df['domestic_gross']
genre_data.head()
```

```python
# obtain the six genres with the most movies

import operator

sum_dict = dict(zip(genre_data.columns, genre_data.sum()))

sorted_dict = sorted(sum_dict.items(), key=operator.itemgetter(1))

sorted_dict[-7:-1]
```

```python
f, axes = plt.subplots(2, 3, figsize=(20, 10), sharex=True)
sns.despine(left=True)

sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Drama'] == 1], palette="Set3", ax=axes[0, 0]).set(ylabel= "Domestic Gross", title = "Drama Movie Success by Month", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Comedy'] == 1], palette="Set3", ax=axes[0, 1]).set(ylabel= "Domestic Gross", title = "Comedy Movie Success by Month", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Action'] == 1], palette="Set3", ax=axes[0, 2]).set(ylabel= "Domestic Gross", title = "Action Movie Success by Month", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Adventure'] == 1], palette="Set3", ax=axes[1, 0]).set(ylabel= "Domestic Gross", title = "Adventure Movie Success by Month", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Thriller'] == 1], palette="Set3", ax=axes[1, 1]).set(ylabel= "Domestic Gross", title = "Thriller Movie Success by Month", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Crime'] == 1], palette="Set3", ax=axes[1, 2]).set(ylabel= "Domestic Gross", title = "Crime Movie Success by Month", xticklabels = tick_label_list)
plt.setp(axes, xticklabels = tick_label_list)

plt.show()
```

```python
f, axes = plt.subplots(2, 3, figsize=(20, 10))
sns.set_context("talk")
# sns.despine(left=True)

sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Drama'] == 1], palette="Set3", ax=axes[0, 0]).set(ylabel= "Domestic Gross", title = "Drama Movie Success by Month \n Total Movies = 347", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Comedy'] == 1], palette="Set3", ax=axes[0, 1]).set(ylabel= "Domestic Gross", title = "Comedy Movie Success by Month \n Total Movies = 257", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Action'] == 1], palette="Set3", ax=axes[0, 2]).set(ylabel= "Domestic Gross", title = "Action Movie Success by Month \n Total Movies = 245", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Adventure'] == 1], palette="Set3", ax=axes[1, 0]).set(ylabel= "Domestic Gross", title = "Adventure Movie Success by Month \n Total Movies = 197", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Thriller'] == 1], palette="Set3", ax=axes[1, 1]).set(ylabel= "Domestic Gross", title = "Thriller Movie Success by Month \n Total Movies = 135", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['Crime'] == 1], palette="Set3", ax=axes[1, 2]).set(ylabel= "Domestic Gross", title = "Crime Movie Success by Month \n Total Movies = 116", xticklabels = tick_label_list)
f.tight_layout()
plt.savefig('Genre_subplots.png')
plt.show()
```

```python
df['release_date'] = df['release_date'].apply(lambda x: pd.to_datetime(x, format = '%Y-%m-%d'))
```

```python
df['year'] = df['release_date'].apply(lambda x: x.year)
```

```python
plt.figure(figsize=(12, 6))
sns.set_context("talk")
ax = sns.boxplot(x="year", y="domestic_gross", data=df, palette="Set3").set(ylabel= "Domestic Gross", title = "Movie Success by Year")
plt.savefig('movie_year_boxplot.png')
plt.show()
```
