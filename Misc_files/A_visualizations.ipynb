```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the .csv we created from the merged datasets
df = pd.read_csv('merged.csv')
```

```python
#Create scatterplot with regression line for average rating vs domestic gross

#sns.regplot(x=df["averagerating"], y=df["domestic_gross"], line_kws={"color":"r","alpha":0.7,"lw":5})
sns.regplot(x=df["averagerating"], y=df["domestic_gross"], fit_reg=True, line_kws={"color":"r","alpha":0.7,"lw":5} ).set(xlabel='Average Rating', ylabel='Domestic Gross', title = 'IMDB Rating vs Domestic Gross')
plt.savefig('rating_gross_scatterplot.png')
plt.show()
```

```python
#Scatterplot of popularity vs domestic gross, with regression line

sns.regplot(x=df["popularity"], y=df["domestic_gross"], line_kws={"color":"r","alpha":0.7,"lw":5})
```

```python
sns.pairplot(df, diag_kind="kde", diag_kws=dict(shade=True, bw=.05, vertical=False) )
# Testing out pairplot for fun. Output 784 figures
```

```python
#Boxplot representing the domestic gross per month
sns.boxplot(x="month", y="domestic_gross", data=df, palette="Greens")
```

```python
#Boxplot representing the average rating per month

sns.boxplot(x="month", y="averagerating", data=df, palette="Set2")
```

```python
#Boxplot representing how movie popularity changes per month
sns.boxplot(x="month", y="popularity", data=df, palette="Set1")
```

```python
tick_label_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
sns.set_context('paper')
ax = sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['studio'] == 'Uni.'], palette="Set2").set(ylabel= "Domestic Gross", title = "Universal Studios Movie Success by Month", xticklabels = tick_label_list)
#plt.savefig('universal_boxplot.png')
plt.show()
```

```python
f, axes = plt.subplots(2, 2, figsize=(25, 15))
sns.set_context("talk")
#sns.despine(left=True)

sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['studio'] == 'Uni.'], palette="Set2", ax=axes[0, 0]).set(ylabel= "Domestic Gross", title = "Universal Studios Movie Success by Month \n Total Movies = 70", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['studio'] == 'Fox'], palette="Set2", ax=axes[0, 1]).set(ylabel= "Domestic Gross", title = "Fox Movie Success by Month \n Total Movies = 61", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['studio'] == 'WB'], palette="Set2", ax=axes[1, 0]).set(ylabel= "Domestic Gross", title = "Warner Bros. Movie Success by Month \n Total Movies = 58", xticklabels = tick_label_list)
sns.boxplot(x="month", y="domestic_gross", data=df.loc[df['studio'] == 'BV'], palette="Set2", ax=axes[1, 1]).set(ylabel= "Domestic Gross", title = "Buena Vista Movie Success by Month \n Total Movies = 46", xticklabels = tick_label_list)

f.tight_layout()
#plt.savefig('Genre_subplots.png')
plt.show()
```
